# How to Convert Text to TF-IDF in Python

What is TF-IDF?

TF-IDF = Term Frequency – Inverse Document Frequency

It gives a score to each word based on:

TF (Term Frequency)
→ How many times a word appears in a document.

IDF (Inverse Document Frequency)
→ How rare or important the word is across all documents.


TfidfVectorizer() → creates TF-IDF tool

fit_transform() → learns words + converts to numbers

Output → matrix of numbers


“Give high score to important words,
Give low score to common words.”

```python
from sklearn.feature_extraction.text import TfidfVectorizer

documents = [
    "I love this movie",
    "This movie is great",
    "I hate this movie"
]

vectorizer = TfidfVectorizer()
tfidf_matrix = vectorizer.fit_transform(documents)

print(tfidf_matrix.toarray())
```

```python
from sklearn.metrics.pairwise import cosine_similarity

cosine_sim = cosine_similarity(tfidf_matrix, tfidf_matrix)
```

```python
movies = movies.head(10000)
```

# Create Title Index Mapping

```python
indices = pd.Series(movies.index, index=movies['title']).drop_duplicates()
```

# Build Recommendation Function

```python
def recommend(title, top_n=5):

    if title not in indices:
        return "Movie not found"

    idx = indices[title]

    sim_scores = list(enumerate(cosine_sim[idx]))

    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)

    sim_scores = sim_scores[1:top_n+1]

    movie_indices = [i[0] for i in sim_scores]

    return movies['title'].iloc[movie_indices]
```

```python
recommend("The Dark Knight")
```

```python
recommend("Toy Story")
```
